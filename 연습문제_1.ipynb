```python
import pandas as pd
df_USA = pd.read_csv('./data/midwest.csv')
```

```python
df_USA
```

```python
df_USA.info()
```

```python
df_USA = df_USA.rename(columns={'poptotal' : 'total', 'popasian' : 'asian'})
```

```python
df_USA
```

```python
df_USA['asian_rate'] = df_USA['asian']/df_USA['total']*100
```

```python
df_USA['asian_rate'].plot.hist()
```

```python
df_USA['mean_result'] = df_USA['asian_rate'].mean()
df_USA['mean_result']
```

```python
import numpy as np
df_USA['result'] = np.where(df_USA['asian_rate'] > df_USA['mean_result'], 'large', 'small')
df_USA['result']
```

```python
count_result = df_USA['result'].value_counts()
count_result.plot.bar()
```
