#### 1. 다음 데이터를 pandas DataFrame으로 만들고, 성별(Gender) 기준으로 나눈 평균 키(Height)를 구하세요.

```python
import pandas as pd
```

```python
DF1 = pd.read_csv('PII.csv')
```

```python
type(DF1)
```

```python
avg_height_by_gender = DF1.groupby('Gender')['Height'].mean()
print(avg_height_by_gender)
```

#### 2. 키(Height)와 몸무게(Weight)를 각각 NumPy 배열로 변환한 후, BMI 지수(BMI = Weight / (Height(m)^2))를 구하여 출력하세요. (단, 키는 m 단위로 변환해야 합니다.)

```python
import numpy as np
```

```python
height = DF1['Height'].values
weight = DF1['Weight'].values
```

```python
height_m = height/100
```

```python
bmi = weight/(height_m **2)
```

```python
DF1['BMI'] = bmi
```

```python
print(DF1[['Height', 'Weight', 'BMI']])
```

#### 3. BloodType이 B형인 학생 중, 몸무게가 70kg 이상인 학생의 이름(Name)을 출력하세요. 

```python
result = DF1[(DF1['BloodType'] == 'B') & (DF1['Weight'] >= 70)]
print(result['Name'])
```

#### 4. 각 성별(Gender)별 평균 몸무게(Weight)를 막대그래프(bar chart) 로 시각화하세요. x축: 성별, y축: 평균 몸무게

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
avg_weight_by_gender = DF1.groupby('Gender')['Weight'].mean().reset_index()
```

```python
plt.figure(figsize = (10, 7))
plt.bar(avg_weight_by_gender['Gender'], avg_weight_by_gender['Weight'], data=avg_weight_by_gender)
plt.title('각 성별 평균 몸무게')
plt.xlabel('성별')
plt.ylabel('평균 몸무게')
plt.show()
```

#### 5. 학년(Grade)별 학생 수를 집계하고, 이를 파이차트(pie chart) 형태로 시각화하세요.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
plt.pie(DF1.Grade.value_counts(),
        labels= DF1.Grade.value_counts().index,
        autopct = '%.1f%%',
        explode = [0.0, 0.0, 0.0, 0.05],
        startangle = 90)
plt.show()
```
